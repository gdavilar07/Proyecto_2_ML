# Notebook 2: Análisis Exploratorio de Datos (EDA)

Este notebook realiza un análisis exploratorio de los datos para entender mejor las características y distribuciones.
Vamos a explorar las distribuciones, correlaciones y detectar valores atípicos.

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
file_path = '../data/raw/HeartDiseaseTrain-Test.csv'
data = pd.read_csv(file_path)

# Distribución de las variables numéricas
data.hist(figsize=(10, 10))
plt.show()

# Correlación entre variables
plt.figure(figsize=(12, 10))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.show()

# Gráficos de cajas para detectar outliers
for column in ['age', 'resting_blood_pressure', 'cholestoral', 'oldpeak']:
    plt.figure()
    sns.boxplot(data[column])
    plt.title(f'Distribución de {column}')
    plt.show()
```
